```python

plast = 0.2 + 0.4*np.linspace(0, 1, 10)
theta = 1.0*np.linspace(0, 1, 10)
```

```python
fig = sns.lineplot(x=np.linspace(0, 1, 10), y=theta, label = "\u03B8" + ", \u03B2 = 1.0")
sns.lineplot(x=np.linspace(0, 1, 10), y=plast, label = "z, x = 0.2, b = 0.4")

sns.set_theme(style='darkgrid', font_scale=0.9)
plt.axvline(x = 0.45, color = 'b', label = "\u03B5" + " = 0.45")
fig.set_xlabel('Environmental Variable, ' + "\u03B5")
fig.set_ylabel('Trait, z and '+ "\u03B8")
leg = plt.legend(loc='upper left')
plt.savefig("rxnNorm_optimal phenotype.png", dpi=500)
```
