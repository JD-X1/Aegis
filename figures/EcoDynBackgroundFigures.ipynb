```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
e = np.linspace(0, 1, 100)
x = np.linspace(0, 30, 100)
sig = 10/(1 + np.exp(-0.35*(x-15)))
linear = np.linspace(0, 10, 100)
log = np.logspace(0, 1, 100)

sns.set_theme(style='darkgrid', font_scale=1.35)
ax = sns.lineplot( y = sig, x = e)
sns.lineplot( y = linear, x = e)
sns.lineplot( y = log, x = e)

#specify axis labels
plt.xlabel('Environment')
plt.ylabel('Phenotype')
plt.tight_layout()
plt.savefig('Reaction_Norms_1.png', dpi=300)
```

```python
e = np.linspace(70, 115, 100)
a = np.linspace(0, 10, 100)
c = np.linspace(3, 7, 100)
b = np.flip(a)
f = np.zeros(100)+5


sns.set_theme(style='darkgrid', font_scale=1.0)
ax = sns.lineplot( y = a, x = e)
sns.lineplot( y = b, x = e)
sns.lineplot( y = c, x = e)
sns.lineplot( y = f, x = e)
ax.set_ylim(0,10)
#specify axis labels
plt.xlabel('Environment (For Example: C' + chr(176)+ ")")
plt.ylabel('Phenotype (Example: Heat Tolerance)')
plt.tight_layout()
plt.savefig('Reaction_Norms_2.png')
plt.show
```
